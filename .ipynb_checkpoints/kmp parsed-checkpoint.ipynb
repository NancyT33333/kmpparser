# Выкачка ссылок

```python
listOfLinks=[]
def findLinks(link):
    page=requests.get(link)
    soup = BeautifulSoup(page.text, 'lxml')
    for i, h in enumerate(soup.findAll("span", {"class" : "pagina" })):
        if(isinstance(h.contents[0], bs4.element.Tag )==True):

            link='https://killpls.me'+h.a["href"]
            if not (link in listOfLinks):
                listOfLinks.append(link)
                print(link)
                findLinks(link)
            
            
    return listOfLinks
findLinks('https://killpls.me/')
        
```

```python
listOfLinks
```

```python
storiesLinks=[]
def getStoriesFromPage(link):
    page = requests.get(link)
    soup = BeautifulSoup(page.text, 'lxml')
    for i, h in enumerate(soup.findAll("a", href=re.compile('^/story.*'))):
        storyLink='https://killpls.me'+h['href']
        if storyLink not in storiesLinks:
            storiesLinks.append(storyLink)
            print('storyLink:'+ storyLink)
    return(storiesLinks)
            
    
```

```python
for link in listOfLinks:
    getStoriesFromPage(link)
storiesLinks
```
