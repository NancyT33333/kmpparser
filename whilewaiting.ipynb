```python
import bs4, re, requests
from bs4 import BeautifulSoup

def getStoryText(link):
    fullStory=[]
    page = requests.get(link)
    soup = BeautifulSoup(page.text,'lxml')
    snippet=soup.find('div', {'class' : "col-xs-12", 'style' : "margin:0.5em 0;line-height:1.785em"})
    for part in snippet.contents:
        if part.string is not None:       
            fullStory.append(part.string.strip())
    fullStoryStr=''.join(fullStory)
    print(link+'\n')
    print(fullStoryStr)
    with open('stories.txt', 'a') as file:
        file.write(fullStoryStr+ "\n")

    return "ok"
```

```python
listOfStories = []
for i in range(24317, 24333):
    listOfStories.append('https://killpls.me/story/' + str(i))
listOfStories[0]
```

```python
for link in listOfStories:
    getStoryText(link)
```
